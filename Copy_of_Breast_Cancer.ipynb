```python
null_rows = df[df.isnull().sum(axis=1) > 0]
print(null_rows)
```

```python
# Calculate the percentage of null values in each column
null_percentages = (df.isnull().sum() / len(df)) * 100
for column in df.columns:
    if null_percentages[column] > 10:
        # Drop the entire column if the percentage of null values is greater than 10%
        df = df.drop(column, axis=1)
    else:
        # Drop null values directly if the percentage is less than or equal to 10%
        df = df.dropna(subset=[column])
```

```python
# Assuming your dataset has a column called 'Surgery_Type' that contains surgery information
surgery_counts = df['Surgery_type'].value_counts()

# Find the most frequent surgery performed
most_frequent_surgery = surgery_counts.idxmax()
frequency = surgery_counts.max()

print("The most frequent surgery performed is {most_frequent_surgery} with {frequency} occurrences.")
```

```python
# Assuming your dataset has a column called 'Surgery_Type' that contains surgery information
surgery_counts = df['Surgery_type'].value_counts()

# Sort the surgery counts in descending order
sorted_surgery_counts = surgery_counts.sort_values(ascending=False)

# Find the most frequent surgery performed
most_frequent_surgery = sorted_surgery_counts.index[0]
frequency = sorted_surgery_counts.iloc[0]

# Find the second most frequent surgery performed
second_most_frequent_surgery = sorted_surgery_counts.index[1]
second_frequency = sorted_surgery_counts.iloc[1]

# Find the third most frequent surgery performed
third_most_frequent_surgery = sorted_surgery_counts.index[2]
third_frequency = sorted_surgery_counts.iloc[2]

print(f"The most frequent surgery performed is {most_frequent_surgery} with {frequency} occurrences.")
print(f"The second most frequent surgery performed is {second_most_frequent_surgery} with {second_frequency} occurrences.")
print(f"The third most frequent surgery performed is {third_most_frequent_surgery} with {third_frequency} occurrences.")
```

```python
# Assuming your dataset has 'Histology' and 'Surgery_Type' columns
# Group the data by 'Histology' and find the most frequent surgery type for each group
prevalent_surgery_by_histology = df.groupby('Histology')['Surgery_type'].agg(lambda x: x.value_counts().index[0])

# Display the prevalent surgery types for each type of breast cancer
print(prevalent_surgery_by_histology)
```

```python
# Assuming your dataset has columns 'Age', 'Protein_1', 'Protein_2', 'Protein_3', and 'Histology'
# Select these columns for the correlation analysis
selected_columns = ['Age', 'Protein1', 'Protein2', 'Protein3', 'Histology']
df_selected = df[selected_columns]

# Calculate the correlation matrix
correlation_matrix = df_selected.corr()

# Create a heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap between Age, Protein Levels, and Breast Cancer Type')
plt.show()
```

```python
# Create subplots for age and protein levels
plt.figure(figsize=(12, 6))

# Box plot for age by Histology
plt.subplot(1, 2, 1)
sns.boxplot(x='Histology', y='Age', data=df)
plt.title('Age Distribution by Histology')

# Box plot for protein levels by Histology
plt.subplot(1, 2, 2)
sns.boxplot(x='Histology', y='Protein1', data=df)
plt.title('Protein_1 Distribution by Histology')

plt.tight_layout()
plt.show()
```
